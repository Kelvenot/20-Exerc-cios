## 7
Leia três notas de um aluno, calcule a média e informe se ele foi aprovado, reprovado ou está em recuperação.

```
notas = []

for i in range(3):
  nota = int(input("Insíra uma nota para o aluno: "))
  notas.append(nota)

media = sum(notas) / len(notas)

print(f"Média final: {media}")

if media >= 6:
  print("Aprovado.")
elif media < 6:
  print("Recuperação.")
```

## 13
Leia um número inteiro e inverta seus dígitos.
Exemplo: 123 → 321

```
entrada = int(input("Insíra um número: "))
soma = 0

while entrada > 0:
  digito = entrada % 10
  soma = (soma * 10) + digito
  entrada //= 10
print(soma)
```

## 16

## 19
Crie um sistema de login com usuário e senha. Permita até 3 tentativas antes de bloquear o acesso.

```
usuarios = []
tentativas = 0
def menu():
  execucao = True
  while execucao and tentativas < 3:
    print("1 - Cadastrar usuário\n2 - Login\n0 - Sair")

    entrada = int(input("Insíra uma opção: "))
    match entrada:
      case 1:
        cadastrar_usuario()
      case 2:
        execucao = login()
      case 0:
        execucao = false
      case _:
        print("Opção inválida.")
  if tentativas >= 3:
    print("Acesso bloqueado.")

def cadastrar_usuario():
  nome = input("Insíra um nome para o usuário.")
  senha = input("Insíra uma senha. ")
  usuario = {
      "nome": nome,
      "senha" : senha
  }
  usuarios.append(usuario)

def login():
  global tentativas

  usuario = input("digite o usuário: ")
  senha = input("digite a senha do usuário: ")

  for user in usuarios:

    if user["nome"] == usuario:

      if user["senha"] == senha:
        print("Login bem sucessido.")
        return False
      else:
        tentativas += 1
        return True
  tentativas += 1
  return True
menu()
```

## 20
Crie um menu interativo com as opções:
1 - Somar
2 - Subtrair
3 - Multiplicar
4 - Dividir
0 - Sair

O programa deve continuar executando até o usuário escolher sair.

```
def menu():
  num = 0
  while True:
    print(f"Número salvo: {num}")
    print("===================================\n1 - Somar\n2 - Subtrair\n3 - Multiplicar\n4 - Dividir\n0 - Sair\n===================================")
    try:
      if num == 0:
        try:
          entrada = int(input("Insíra o número para ser calculado: "))
        except ValueError:
          print("Por favor insíra um número.")
      else:
        entrada = num

      opcao = int(input("Qual opção deseja acessar: "))

      match opcao:
        case 1:
          num = somar(entrada)
        case 2:
          num = subtrair(entrada)
        case 3:
          num = multiplicar(entrada)
        case 4:
          num = dividir(entrada)
        case 0:
          print(f"O resultado é {entrada}")
          break
        case _:
          print("Opção inválida.")

    except ValueError:
      print("Por favor insíra um número.")
```

```
def somar(numero):
  try:
    entrada = int(input(f"Insíra o número para ser somado a {num}: "))
  except ValueError:
    print("Por favor insíra um número.")

  valor = entrada + numero
  return valor
```

```
def subtrair(numero):
  try:
    entrada = int(input(f"Insíra o número para ser subtraido de {num}: "))
  except ValueError:
    print("Por favor insíra um número.")

  valor = numero - entrada
  return valor
```

```
def multiplicar(numero):
  try:
    entrada = int(input(f"Insíra o número para multiplicar {num}: "))
  except ValueError:
    print("Por favor insíra um número.")

  valor = numero * entrada
  return valor
```

```
def divididir(numero):
  try:
    entrada = int(input(f"Insíra o número para ser dividir {num}: "))
  except ValueError:
    print("Por favor insíra um número.")

  valor = numero / entrada
  return valor
```

```
menu()
```
